```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("thedevastator/weather-prediction")
path=os.path.join(path,"weather_prediction_dataset.csv")
ds=pd.read_csv(path)
ds
```

```python
ds.columns
```

```python
ds['mean_temp_all']=ds.filter(like="_temp_mean").mean(axis=1)
```

```python
ds
```

#Lets consider running the model to find out the mean_temp of BASEL 

```python
index=0
start_ind=index*10
features_basel=ds.iloc[:, start_ind:start_ind+10]
features_basel=features_basel.drop(columns=['DATE','MONTH'])
target= features_basel['BASEL_temp_mean']
features=features_basel.drop(columns=['BASEL_temp_mean'])
```

```python
features.columns
```

```python
features.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(features.values,target.values,test_size=0.3, random_state=45)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
X_train_tensor=torch.tensor(X_train,dtype=torch.float32)
X_test_tensor=torch.tensor(X_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.float32).unsqueeze(1)
y_test_tensor=torch.tensor(y_test,dtype=torch.float32).unsqueeze(1)
```

```python
X_train_tensor.shape
```

```python
y_train_tensor
```

```python
class WeatherDataset(Dataset):
    def __init__(self,features,target):
        self.features=features
        self.target=target
    def __len__(self):
        return len(self.features)
    def __getitem__(self, ind):
        return self.features[ind], self.target[ind]
```

```python
train_ds=WeatherDataset(X_train_tensor,y_train_tensor)
test_ds=WeatherDataset(X_test_tensor,y_test_tensor)

train_dataloader=DataLoader(train_ds, batch_size=32, shuffle=True)
test_dataloader=DataLoader(test_ds, batch_size=32, shuffle=True)
```

```python
train_dim=X_train.shape[1]
```

```python
import torch.nn as nn
class Model(nn.Module):
    def __init__(self,input_dim):
        super(Model,self).__init__()
        self.fc1=nn.Linear(input_dim,82)
        self.fc2=nn.Linear(82,41)
        self.fc3=nn.Linear(41,1)
    def forward(self,output):
        output=torch.relu(self.fc1(output))
        output=torch.relu(self.fc2(output))
        output=self.fc3(output)
        return output
```

```python
weath_model=Model(train_dim)
weath_model
```

```python
import torch.optim as optim
crit=nn.MSELoss()
optimizer=optim.Adam(weath_model.parameters(), lr=0.001)
```

```python
epochs=1000
for epoch in range(epochs):
    weath_model.train()
    train_loss=0
    for X_batch, y_batch in train_dataloader:
        optimizer.zero_grad()
        preds=weath_model(X_batch)
        loss=crit(preds,y_batch)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {train_loss/len(train_dataloader):.4f}")
```

```python
weath_model.eval()
```

```python
torch.save(weath_model.state_dict(),"weather_model.pt")
```

```python
test_loss = 0
with torch.no_grad():
    for X_batch, y_batch in test_dataloader:
        preds = weath_model(X_batch)
        loss = crit(preds, y_batch)
        test_loss += loss.item()
        
avg_test_loss = test_loss / len(test_dataloader)
print(f"Test Loss (MSE): {avg_test_loss:.4f}")
total_variance = ((y_batch - y_batch.mean()) ** 2).sum()
residual_variance = ((y_batch - preds) ** 2).sum()
r_squared = 1 - (residual_variance / total_variance)
print(f"R²: {r_squared:.4f}")
```
